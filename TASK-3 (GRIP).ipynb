```python
df['Sub-Category'].value_counts()    #Counts the individual values of different Sub-Categories
```

```python
#Plotting the above values on a pie chart
plt.figure(figsize=(10,10))
df['Sub-Category'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.show()
```

### Note 1: Maximum purchases are for binders, paper, furnishings, phones, storage, etc. and minimum purchases are for copiers, machines, suppliers,fasteners, etc.

```python
df['State'].value_counts()   #Counts the buyers from different states
```

```python
#Plotting the above values on a bar chart
plt.figure(figsize=(15,12))
sns.countplot(x='State',data=df,palette='tab10',order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

### Note 2: Maximum number of buyers are from California, New York and Texas.

```python
df.hist(figsize=(10,10),bins=50)
plt.show()
```

## Note 3: 
#### (i) Most customers tend to buy quantities of 2 and 3
#### (ii) Maximum discount offered is between 0 to 20 percent

```python
#Plotting the values in accordance with sub-region: East, West, Central & South
plt.figure(figsize=(10,8))
df['Region'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.show()
```

### Comparison between Profit and Sales:

```python
#Plotting the chart for Profit vs Sales
fig,ax=plt.subplots(figsize=(20,8))
ax.scatter(df['Sales'],df['Profit'])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

### Comparison between Profit and Discount

```python
#Plotting the chart for Profit vs Discount
sns.lineplot(x='Discount',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

### Comparison between Profit and Quantity

```python
#Plotting the chart for Profit vs Quantity
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

```python
#Plotting the sales and profit for each segment
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['lime','red'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

<b> Profit and sales are maximum in Consumer Segment and minimum in Home Office segment. <b>

```python
#Plotting the sales of different segments according to the regions
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index,palette='Spectral')
plt.xlabel('Region',fontsize=15)
plt.show()
```

```python
#Plotting Profit and Sales distribution, region-wise 
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['lime','red'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

<b> Profit and Sales are maximum in west region and minimum in south region. <b>

```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['lime','red'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

> **California, New York, Washington, etc.** have high profits <br>
> **Texas, Pennsylvania, Ohio & 7 more** made losses

```python
t_states = df['State'].value_counts().nlargest(10)
t_states
```

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['lime','red'],alpha=0.9,figsize=(10,8))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

<b> Technology and Office Supplies have higher profits whereas Furniture has less profits. <b>

```python
ps = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['lime','red'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

### Conclusion:
1. Phones category has the highest sales
2. Chairs have high sales but comparitively less profit than phones.
3. Tables and Bookcases category are facing losses.
4. Envelopes, Labels and Fastners contribute the least towards sales and profit.
5. Paper, Copiers and Envelopes have higher profit margin.
